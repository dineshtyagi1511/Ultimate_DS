```python
import numpy as np 
```

The Model we are using is a Multiple linear Regression(MLR) model , and the optimization method we are apllying is a Gradient Descent
 


```python
##DATA 
X = np.array([[1,1,0],
             [1,2,1],
             [1,3,1],
             [1,4,2],
             [1,5,3],
             [1,6,4]])  # 6*3
y= np.array([39.6,47.100,56.978,68.240,77.867,85.022])


```

```python
## Initialize 
beta =np.zeros(3) ## b0,b1,b2
learning_rate = 0.01
m = len(y)
print("Initialize Beta", beta)
```

```python
## Epoch 1
print("===Epoch===")

## Step 1:Compute Prediction 
y_pred = np.dot(X,beta.T)
print("predictions:",y_pred)

## step 2:Compute errors
errors= y-y_pred
print("Errors:",errors)

## step 3:Compute gradient 
gradient = (np.dot(X.T,(y_pred-y)))/m
print("Gradient:",gradient)

## step 4:update beta :
beta = beta - learning_rate*gradient
print("Updated beta:",beta)


y_pred = np.dot(X,beta)
print("Predictions:",y_pred)
```

```python
## Epoch 2
print("===Epoch2===")

## Step 1:Compute Prediction 
y_pred = np.dot(X,beta.T)
print("predictions:",y_pred)

## step 2:Compute errors
errors= y-y_pred
print("Errors:",errors)

## step 3:Compute gradient 
gradient = (np.dot(X.T,(y_pred-y)))/m
print("Gradient:",gradient)

## step 4:update beta :
beta = beta - learning_rate*gradient
print("Updated beta:",beta)


y_pred = np.dot(X,beta)
print("Predictions:",y_pred)
```

```python
for epoch in range(3,1000):
    print(f"====Epoch{epoch}====")

    
    ## Step 1:Compute Prediction 
    y_pred = np.dot(X,beta.T)
    print("predictions:",y_pred)

    ## step 2:Compute errors
    errors= y-y_pred
    print("Errors:",errors)

    ## step 3:Compute gradient 
    gradient = (np.dot(X.T,(y_pred-y)))/m
    print("Gradient:",gradient)

    ## step 4:update beta :
    beta = beta - learning_rate*gradient
    print("Updated beta:",beta)


    y_pred = np.dot(X,beta)
    print("Predictions:",y_pred)
```
